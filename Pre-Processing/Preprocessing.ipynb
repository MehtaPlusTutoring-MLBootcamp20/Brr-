```
file_path = "/content/drive/My Drive/MIT AI Class Work/Final Project"
file_path_true = file_path + "/True.csv"
file_path_fake = file_path + "/Fake.csv"
```

```
true_df2 = true_df.drop(columns = ["date", "subject"])
fake_df2 = fake_df.drop(columns = ["date", "subject"])
```

```
true_df2.head()
```

```
fake_df2.head()
```

```
zero_data = np.zeros(shape=(len(true_df2),1))
d_True = pd.DataFrame(zero_data, columns=["Is_Fake"])
d_True.head()
```

```
ones_data = np.ones(shape=(len(fake_df2),1))
d_Fake = pd.DataFrame(ones_data, columns=["Is_Fake"])
d_Fake.head()
```

```
true_df3 = pd.concat([true_df2, d_True], axis=1)
true_df3.head()
```

```
fake_df3 = pd.concat([fake_df2, d_Fake], axis=1)
fake_df3.head()
```

```
fake_df3["title"].value_counts()
```

```
fake_df3["text"].value_counts()
```

```
df = pd.concat([fake_df3, true_df3], axis=0)
df.head()
```

```
df["Is_Fake"].value_counts()
```

```
df.loc[df['text'] == " ", 'text']= np.nan
df.loc[df['text'] == "  ", 'text']= np.nan
```
